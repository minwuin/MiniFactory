```python
import os
import shutil
import random

source_dir = r"C:\minwoin\miniproject\SmartFactory\pc\factoryimage"
dataset_dir = r"C:\minwoin\miniproject\SmartFactory\pc\dataset"

def make_dirs():
    folders = [
        "images/train", "images/val",
        "labels/train", "labels/val"
    ]
    for folder in folders:
        os.makedirs(os.path.join(dataset_dir, folder), exist_ok=True)

make_dirs()

image_files = [f for f in os.listdir(source_dir) if f.endswith(('.jpg', '.jpeg', '.png'))]
random.shuffle(image_files)

split_index = int(len(image_files) * 0.8)
train_files = image_files[:split_index]
val_files = image_files[split_index:]

def move_files(file_list, split_type):
    for img_name in file_list:
        base_name = os.path.splitext(img_name)[0]
        txt_name = base_name + ".txt"
        
        img_src = os.path.join(source_dir, img_name)
        txt_src = os.path.join(source_dir, txt_name)
        
        if os.path.exists(txt_src):
            shutil.copy(img_src, os.path.join(dataset_dir, "images", split_type, img_name))
            shutil.copy(txt_src, os.path.join(dataset_dir, "labels", split_type, txt_name))

move_files(train_files, "train")
move_files(val_files, "val")

# 핵심 수정: 클래스 개수를 9로 변경하고 목록 맨 앞에 'ground' 추가
yaml_content = f"""train: {os.path.join(dataset_dir, 'images', 'train')}
val: {os.path.join(dataset_dir, 'images', 'val')}

nc: 9
names: ['ground', 'car', 'A', 'B', 'C', 'D', 'E', 'F', 'G']
"""

with open(os.path.join(dataset_dir, "data.yaml"), "w", encoding="utf-8") as f:
    f.write(yaml_content)

print(f"데이터셋 구조화가 완료되었습니다. 학습 이미지: {len(train_files)}장, 검증 이미지: {len(val_files)}장")
```

```python
from ultralytics import YOLO

# 매개변수가 가장 많은 대형 모델로 변경
model = YOLO("yolo26l-seg.pt")

model.train(
    data=r"C:\minwoin\miniproject\SmartFactory\pc\dataset\data.yaml",
    epochs=300,       # 반복 학습 횟수 증가
    patience=50,      # 오십 번 동안 성능 개선이 없으면 조기 종료
    batch=4,         # 그래픽 카드 성능에 맞춰 일괄 처리 데이터 크기 자동 최대화
    imgsz=640,        # 이미지 해상도
    device=0,         # 그래픽 카드 연산 장치 번호
    project="factory_fms",
    name="v3_segmentation",
    task="segment"    # 컴퓨터에게 분할 작업을 수행함을 명시
)
```
